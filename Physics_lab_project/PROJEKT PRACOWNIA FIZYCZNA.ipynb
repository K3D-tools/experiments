Struktura materii może być modelowana za pomocą $N$ elementowego zbioru punktów w przestrzeni trójwymiarowej.  

W przypadku gazów punkty reprezentują obecne położenia atomów (albo jonów lub molekuł) natomiast dla kryształów są to punkty równowagi wokół, których atomy (albo jony lub molekuły) wykonują drgania. 

Modele gazów zakładają brak oddziaływań międzyatomowych co sprawia, że ułożenie atomów względem siebie w przestrzeni jest przypadkowe, za to między atomami w krysztale występują wiązania co narzuca pewne uporządkowanie. 

By opisać model struktury tego typu dla materii w stanie gazowym należy określić położenia wszystkich atomów, a do odtworzenia modelowej struktury kryształu wystarczy komórka elementarna wraz z wektorami translacji.


Przykładowo dla $N$ elementowej, krystalicznej sieci regularnej wystarczy położenie równowagi jednego atomu oraz $N-1$ przesunięć wraz z duplikacją o wektor:

$$\boldsymbol{T} = h\boldsymbol{a} + k\boldsymbol{b} + l\boldsymbol{c} \qquad h,k,l \in \mathbb{Z}$$

Cechą ciała krystalicznego jest uporządkowanie bliskiego i dalekiego zasięgu, natomiast w przypadku gazu położenia cząstek są zupełnie nieskorelowane.

W przypadku kryształu, każdy węzeł sieci posiada identyczne otoczenie. 

Dla każdego węzła sieci w szczególnej odległości, znajduje się dana ilość sąsiadujących węzłów. 

Przykładowo każdy atom sieci regularnej z ilustracji posiada dokładnie $6$ najbliższych sąsiadów w odległości $|a|=|b|=|c|$ (na rysunku linie koloru niebieskiego i zielonego). Kąty wiązań utworzone przez linie zbiegające się w jednym punkcie są równe $\alpha=\beta=\gamma = arccos(\frac{ \boldsymbol{b} \cdot \boldsymbol{c} }{ | \boldsymbol{b} | \cdot | \boldsymbol{c} |})= arccos(\frac{ \boldsymbol{a} \cdot \boldsymbol{c} }{ | \boldsymbol{a} | \cdot | \boldsymbol{c} |})= arccos(\frac{ \boldsymbol{a} \cdot \boldsymbol{b} }{ | \boldsymbol{a} | \cdot | \boldsymbol{b} |}) = \frac{\pi}{2}$. 

Te zależności oraz $8$ sąsiadów w odległości $|a| \sqrt{3}=|b| \sqrt{3}=|c| \sqrt{3}$ (na rysunku linie koloru czerwonego i magenta) są przejawami istnienia korelacji w lokalnym ułożeniu atomów.

Wyżej wspomniana liczba najbliższych sąsiadów w krysztale nazywana bywa liczbą koordynacyjną. 

Pojedynczą liczbę koordynacyjną można uogólnić na sekwencję liczb określających liczbę sąsiadów w kolejnych warstwach otaczających atom. 

Ten sposób opisywania struktury będący zależnością liczby sąsiadów od odległości od punktu odniesienia nazywany jest radialną funkcją rozkładu (radial distribution function).

W przypadku modelu idealnego kryształu radialna funkcja rozkładu jest dyskretną sekwencją warstw koordynacyjnch. 

Przykład przedstawia zależność liczby koordynacyjnej ($z$) od promienia strefy warstwy koordynacyjnej dla regularnej sieci ściennie centrowanej.

W odległości $r = 0.5 \sqrt{2}$ znajduje się $z(r) = 12$, a dla $r = 1 $ znajduje się $z(r) = 6$ sąsiadów.
